```python
# Now identifying the categorical coloumns and converting them into categorical data as they are already in numerical type.
```

```python
df["Reason for absence"] = df["Reason for absence"].apply(lambda x : str(x) )
```

```python
df.info()
```

```python
df["Month of absence"] = df["Month of absence"].apply(lambda x : str(x))
df["Day of the week"] = df["Day of the week"].apply(lambda x : str(x))
df["Seasons"] = df["Seasons"].apply(lambda x : str(x))
df["Age"] = df["Age"].apply(lambda x : str(x))
df["Disciplinary failure"] = df["Disciplinary failure"].apply(lambda x : str(x))
df["Education"] = df["Education"].apply(lambda x : str(x))
df["Social drinker"] = df["Social drinker"].apply(lambda x : str(x))
df["Social smoker"] = df["Social smoker"].apply(lambda x : str(x))
```

```python
df.info()
```

```python
df["Work load Average/day "] = df["Work load Average/day "].apply(lambda x : str(x).replace("," , "")  if "," in str(x) else str(x) )
```

```python
#Now "Work load Average/day " feature is categorical type but its actually Numerical type so before converting to float, first removing the comas and converting it into float.
```

```python
df.head()
```

```python
df["Work load Average/day "] = df["Work load Average/day "].apply(lambda x : float(x))
```

```python
#Converting "Work load Average/day " into float
```

```python
df.dtypes
```

```python
#Above Output Shows that "Work load Average/day " has been converted to float
```

```python
#Now below code :  NANâ€™s for categorical columns to be replaced with its previous row values
```

```python
col = ["Reason for absence","Month of absence","Day of the week","Seasons","Age","Disciplinary failure","Work load Average/day ","Education","Social drinker","Social smoker"]
```

```python
df.loc[:,col] = df.loc[:,col].ffill()
```

```python
numeric_col_discrete = ["Son","Pet"] 
numeric_col_continous = ["Transportation expense","Distance from Residence to Work","Service time","Hit target","Weight","Height","Body mass index","Absenteeism time in hours"]
```

```python
df["Son"]=df["Son"].fillna(math.floor(df["Son"].mean()))

for i in numeric_col_continous:
    df[i]=df[i].fillna(df[i].mean())
```

```python
#Above Code is to replace the missing numerical values with mean of it
```

```python
df.isnull().sum()
```

```python
#Above table shows that there are no missing values.
```

```python
df.head(100)
```

```python
age_work_sum = df.groupby('Age', as_index=False)[['Work load Average/day ']].mean()
ax = age_work_sum.plot(kind='bar', x='Age', figsize=(8,6))
ax.set_ylabel('Work load Average/day')
ax.set_title('Average work load/day by age')
plt.show()
```

```python
#Absenteeism seems to be same across Age 
```

```python
#Average Absenteeism hours by Distance to work
# 1st Hypothesis
# HYPOTHESES : distance to work increases the absenteeism hours.
```

```python
dis_abs = df.groupby('Distance from Residence to Work')[['Absenteeism time in hours']].mean()
ax = dis_abs.plot(kind='bar', figsize=(8,6), legend=False)
for i, v in enumerate(dis_abs.values):
    ax.text(i-.25, v + 1, str(np.int(np.round(v))), color='red')
ax.set_xlabel('Distance from Residence to Work (km)')
ax.set_ylabel('Absenteeism time in hours')
ax.set_title('Average Absenteeism time in hours by distance')
plt.show()
```

```python
#From the above graph, we see that our HYPOTHESES is not correct.
```
